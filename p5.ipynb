```python
import hashlib, Crypto.PublicKey.RSA

def UAB_h(m):
    return int(hashlib.sha1('%s' % m).hexdigest(), 16)

def UAB_xor(a, b):
    return eval("%s^%s" % (a, b))

#Returns a tuple with the (n, e) values of a new public key
def UAB_generate_RSA_public_key(nBits):
    keys = Crypto.PublicKey.RSA.generate(nBits, os.urandom)
    pk = keys.publickey()
    return (pk.n, pk.e)

#Returns a tuple ((n, e), (n, d)) containing a public key tuple in the 1st position and a private key tuple in the 2nd position.
def UAB_generate_RSA_key_pair(nBits):
    keys = Crypto.PublicKey.RSA.generate(nBits, os.urandom)
    pk = keys.publickey()
    pk_vars = (pk.n, pk.e)
    sk_vars =  (keys.n, keys.d)
    return(pk_vars, sk_vars)
```

```python
# EXERCISE 2a: Implement a function to obtain the y value for the signer by solving equation C_k_v(y1, y2, ..., yr) == v.
#
# Function UAB_solve_C.
# * Parameter k: key used in the symmetric crypto system
# * Parameter v: initialization vector
# * Parameter ys: random inputs. Being signer user i, y[i] = None. E.g. if in a 6 user array, signer is in position 3, then ys = [y0,y1,y2,None,y4,y5]
# * Returns: y value of the signer

def UAB_solve_C(k,v,ys):
    res = None
    #### IMPLEMENTATION GOES HERE ####
    a=v
    pos=0
    #apliquem la funció de combinació en sentit horari fins que trobem la y que volem calcular
    for x,yi in enumerate(ys):
        if(yi != None):
            
            a=UAB_xor(a,yi)
            a=UAB_E(a,k)
        else:
            pos=x
            break
    b=v
     #apliquem la funció de combinació en sentit antihorari fins que trobem la y que volem calcular
    for yi in reversed(ys):
        if(yi != None):
            b=UAB_xor(b,k)
            b=UAB_E(b,yi)
        else:
            break
    b=UAB_E(b,k)# ens queda aplicar una k en l'esquema
    res=UAB_xor(a,b)# amb els valors que hem trobat fem una xor per trobar y 
    ##################################
    return res
```

```python
# EXERCISE 2b: Implement the UAB_sign_ring_simple function.
#
# Function UAB_sign_ring_simple.
# * Parameter m: message to sign.
# * Parameter v: initialization vector.
# * Parameter xs: Random values for each user. Being signer user i, x[i] = None. E.g. if in a 6 user array, signer is in position 3, then xs = [x0,x1,x2,None,x4,x5]
# * Parameter publicKeysGroup: array with the public keys of the members of the group.
# * Parameter skSigner: private key of the signer.
# * Returns: ring signature sigma. Sigma is a tuple with three elements:
#    1. List of public keys of the group members.
#    2. Initialization vector v.
#    3. List of the x values of all the members.

def UAB_sign_ring_simple(m, v, xs, publicKeysGroup, skSigner):
    res = None
    #### IMPLEMENTATION GOES HERE ####
    k=UAB_h(m)
    #print(k)
    #print("\n")
    ys= [None] * len(xs)
    xaux=[]
    #no es necessario si se ejecuta cell->Run All. Pero si se ejecuta solamente una celda el valor de xs se ha sobreescrito
    # y da un resultado erroneo en los siguientes intentos.
    for element in xs:
        xaux.append(element)
    pos=0
    for i,x in enumerate(xs):
        if x != None:
            
            ys[i]=UAB_f(publicKeysGroup[i],x)#calculem les diferents y a partir de les x, excepte pel que signa.
        else:
            pos=i
            
    y=UAB_solve_C(k,v,ys)# trobem el valor de y del signant a partir de la funció de combinació
    x=UAB_f_inv(skSigner,y)# trobem la x
    xaux[pos]=x #assignem la x a la posició on es trobra el None,ja tenim totes les xs
    res= (publicKeysGroup,v,xaux)
    ##################################
    return res
```

```python
# EXERCISE 3a: Implement the UAB_verify_ring_signature_simple function
#
# Function UAB_verify_ring_signature_simple.
# * Parameter m: message that has been signed.
# * Parameter sigma: ring signature. This is a tuple with three elements: 
#    1. List of public keys of the group members.
#    2. Initialization vector v.
#    3. List of the x values of all the members.
# * Returns: a boolean value indicating if sigma is a valid ring signature for the message m.

def UAB_verify_ring_signature_simple(m, sigma):
    res = None
    #### IMPLEMENTATION GOES HERE ####
    k=UAB_h(m)
    r=sigma[1] #inicialitem r amb v
    for i,signatura in enumerate(sigma[0]):
        x=sigma[2][i]
        y=UAB_f(signatura,x)#calculem les y a partir de les x
        r=UAB_xor(r,y)#fem la xor entre la y i el valor anterior     
        r=UAB_E(r,k) #fem la xor entre el valor anterior i k
    res= (r == sigma[1])#comprovem si es compleix que el resultat de la funció de combinació és igual a v
    ##################################
    return res
```

```python
# EXERCISE 4c: Implement the UAB_sign_ring function and UAB_verify_ring_signature.
#
# Function UAB_sign_ring.
# * Parameter m: message to sign.
# * Parameter v: initialization vector. 
# * Parameter xs: Random values for each user. Being signer user i, x[i] = None. E.g. if in a 6 user array, signer is in position 3, then xs = [x0,x1,x2,None,x4,x5]
# * Parameter publicKeysGroup: array with the public keys of the members of the group.
# * Parameter skSigner: private key of the signer.
# * Returns: ring signature sigma. Sigma is a tuple with three elements:
#    1. List of public keys of the group members.
#    2. Initialization vector v.
#    3. List of the x values of all the members.

def UAB_sign_ring(m, v, xs, publicKeysGroup, skSigner):
    res = None
    
    #### IMPLEMENTATION GOES HERE ### 
    k=UAB_h(m)
    ys= [None] * len(xs)
    ni=[]
    for element in publicKeysGroup:
        ni.append(element[0])
        
    max_ni=max(ni)
    keep=True
    b=0
    #calculem el valor de b
    while(keep):
        
        if(pow(2,b)>max_ni):
            keep=False
        else:
            b=b+1
    xaux=[]
    #no es necessario si se ejecuta cell->Run All. Pero si se ejecuta solamente una celda el valor de xs se ha sobreescrito
    # y da un resultado erroneo en los siguientes intentos.
    for element in xs:
        xaux.append(element)
    
    pos=0
    for i,x in enumerate(xs):
        if x != None:
            ys[i]=UAB_g(publicKeysGroup[i],x,b)#calculem les y
        else:
            pos=i       
    y=UAB_solve_C(k,v,ys)#calculem la y del signador
    x=UAB_g(skSigner,y,b)#calculem la x del signador
    xaux[pos]=x#asignem la x del signador a la posició None
    res=(publicKeysGroup,v,xaux)#retornem la signatura
    ##################################
    
    return res

# Function UAB_verify_ring_signature.
# * Parameter m: message that has been signed.
# * Parameter sigma: ring signature. This is a tuple with three elements:
#    1. List of public keys of the group members.
#    2. Initialization vector v.
#    3. List of the x values of all the members.
# * Returns: a boolean value indicating if sigma is a valid ring signature for the message m.

def UAB_verify_ring_signature(m, sigma):
    res = None
    
    #### IMPLEMENTATION GOES HERE ####
    ni=[]
    for element in sigma[0]:
        ni.append(element[0])
        
    max_ni=max(ni)
    keep=True
    b=0
    
    #trobem b
    while(keep):
        if(pow(2,b)>max_ni):
            keep=False
        else:
            b=b+1
    
    k=UAB_h(m)
    r=Integer(sigma[1]) #inicialitem r amb v
    for i,signatura in enumerate(sigma[0]):
        x=Integer(sigma[2][i])
        y=Integer(UAB_g(signatura,x,b))#trobem la y
        r=UAB_xor(r,y)#calculem la xor del pas anterior amb y
        r=UAB_xor(r,k)#xor del pas anterior amb k
    res = (r == sigma[1])#comprovem que el resultat de la funció de combinació sigui v
    ##################################
    
    return res
```
